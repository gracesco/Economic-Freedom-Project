```python
import pandas as pd
```

```python
# Create a path for the CSV file 
csv_path = "cleanData.csv"

# Read the CSV into a Pandas DataFrame
cleanData_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
cleanData_df.head()
```

```python
OriginalGDP_df = cleanData_df.loc[:,["Country Name", "GDP ($, Billions, PPP)", "Public Debt (% of GDP)", "Gov't Expenditure (% of GDP)"]]
OriginalGDP_df = OriginalGDP_df.rename(columns={"Country Name": "Country", "GDP ($, Billions, PPP)": "Avg_GDP($,B,PPP)", "Public Debt (% of GDP)": "Avg_Public_Debt(%)", "Gov't Expenditure (% of GDP)": "Avg_Gov_Expenditure(%)"})
OriginalGDP_df.style.format({"Avg_GDP($,B,PPP)": "$ {:.2f}", "Avg_Public_Debt(%)": "{:.2f}", "Avg_Gov_Expenditure(%)": "{:.2f}"})
```

```python
RegionalGDP_df = cleanData_df.loc[:,["Country Name", "Region"]]
RegionalGDP_df = RegionalGDP_df.groupby("Region", as_index=False)
RegionalGDP_df = RegionalGDP_df.count()
RegionalGDP_df = RegionalGDP_df.rename(columns={"Country Name": "#_of_Countries"})

GDP_df = cleanData_df.loc[:,["Region", "GDP ($, Billions, PPP)", "Public Debt (% of GDP)", "Gov't Expenditure (% of GDP)"]]
GDP_df = GDP_df.groupby("Region", as_index=False).mean()

CleanGDP_df = GDP_df.merge(RegionalGDP_df, on="Region")
CleanGDP_df = CleanGDP_df.rename(columns={"GDP ($, Billions, PPP)": "Avg_GDP($,B,PPP)", "Public Debt (% of GDP)": "Avg_Public_Debt(%)", "Gov't Expenditure (% of GDP)": "Avg_Gov_Expenditure(%)"})
CleanGDP_df.style.format({"Avg_GDP($,B,PPP)": "$ {:.2f}", "Avg_Public_Debt(%)": "{:.2f}", "Avg_Gov_Expenditure(%)": "{:.2f}"})
```

```python
calculate_df = cleanData_df.loc[:,["Region","Population (Millions)","Unemployment (%)","GDP ($, Billions, PPP)", "Public Debt (% of GDP)", "Gov't Expenditure (% of GDP)", ]]

calculate_df["Avg Unemployment"] = (calculate_df["Unemployment (%)"]/100) * calculate_df["Population (Millions)"] 
calculate_df["Avg Public Debt"] = (calculate_df["Public Debt (% of GDP)"]/100) * calculate_df["GDP ($, Billions, PPP)"] 
calculate_df["Avg Gov't Expenditure"] = (calculate_df["Gov't Expenditure (% of GDP)"]/100) * calculate_df["GDP ($, Billions, PPP)"] 
calculate_df = calculate_df.groupby("Region", as_index=False).mean()
calculate_df = calculate_df.drop(columns=["Public Debt (% of GDP)", "Gov't Expenditure (% of GDP)", "Unemployment (%)"],)
calculate_df = calculate_df.rename(columns={"GDP ($, Billions, PPP)": "Avg GDP ($,B,PPP)","Population (Millions)": "Population (M)","Avg Unemployment": "Avg Unemployment (M)", "Avg Public Debt": "Avg Public Debt($,B)", "Avg Gov't Expenditure" : "Avg Gov Expenditure ($,B)"})


calculate_df.head()
```

```python
OriginalGDP_df.to_csv(r'GDPStatsbyCountry.csv') 
CleanGDP_df.to_csv(r'GDPStatsbyRegion.csv')
calculate_df.to_csv(r'UnemploymentGDP.csv')
```
